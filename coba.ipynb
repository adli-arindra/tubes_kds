```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq

# Given parameters
Fs = 250  # Sampling frequency (Hz)
Ts = 1 / Fs  # Sampling period (s)
T_max = 0.1  # Time range for plotting (s)
N = int(T_max * Fs)  # Number of samples

# Time vectors
t_cont = np.linspace(0, T_max, 1000)  # Continuous time for analog signal
n = np.arange(0, N)  # Discrete time indices
t_disc = n * Ts  # Discrete time values

# Define analog and sampled signals
xa_t = 5 + 4 * np.sin(50 * np.pi * t_cont + np.pi/4) + 3 * np.cos(250 * np.pi * t_cont)
x_n = 5 + 4 * np.sin(50 * np.pi * t_disc + np.pi/4) + 3 * np.cos(250 * np.pi * t_disc)
```

```python
# Plot Analog Signal
plt.figure(figsize=(10, 5))
plt.plot(t_cont, xa_t, label="Analog Signal $x_a(t)$", color='b')
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Sampled Signal
plt.figure(figsize=(10, 5))
plt.stem(t_disc, x_n, linefmt='r-', markerfmt='ro', basefmt="k-", label="Sampled Signal $x(n)$")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Analog Signal
plt.figure(figsize=(10, 5))
plt.plot(t_cont, xa_t, label="Analog Signal $x_a(t)$", color='b')
plt.stem(t_disc, x_n, linefmt='r-', markerfmt='ro', basefmt="k-", label="Sampled Signal $x(n)$")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Compute DTFT using FFT approximation
X_w = fft(x_n)  # Compute FFT
frequencies = fftfreq(N, Ts)  # Frequency axis in Hz

# Plot DTFT Magnitude Spectrum
plt.figure(figsize=(10, 5))
plt.plot(frequencies[:N//2], np.abs(X_w[:N//2]), label="Magnitude Spectrum", color='g')
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.legend()
plt.show()
```
